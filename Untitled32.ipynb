```python
import minari
import torch
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
```

```python
# -----------------------------
# Load dataset (with automatic download)
# -----------------------------
dataset_name = 'mujoco/halfcheetah/simple-v0'  # Change to 'medium', 'expert', etc.
dataset = minari.load_dataset(dataset_name, download=True)  # <-- download=True

# Recover the environment if needed
env = dataset.recover_environment()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
```

```python
# -----------------------------
# Print basic information about the dataset object
# -----------------------------
print("Type of dataset object:", type(dataset))
print("\n--- Attributes ---")
print([attr for attr in dir(dataset) if not attr.startswith("_")])

# Optionally, list callable methods
print("\n--- Methods ---")
methods = [method for method in dir(dataset) if callable(getattr(dataset, method)) and not method.startswith("_")]
print(methods)
```

```python
# -----------------------------
# Sample episodes and align obs/actions
# -----------------------------
episodes = dataset.sample_episodes(dataset.total_episodes)
obs_list, act_list = [], []

for ep in episodes:
    # Align obs with actions: ignore last observation
    obs_list.append(ep.observations[:-1])
    act_list.append(ep.actions)

# Concatenate all episode data
observations = torch.tensor(np.concatenate(obs_list, axis=0), dtype=torch.float32)
actions = torch.tensor(np.concatenate(act_list, axis=0), dtype=torch.float32)

# -----------------------------
# Create PyTorch Dataset & DataLoader
# -----------------------------
tensor_dataset = TensorDataset(observations, actions)
loader = DataLoader(tensor_dataset, batch_size=64, shuffle=True)

# -----------------------------
# Define Behavior Cloning Policy (MLP)
# -----------------------------
class BCPolicy(nn.Module):
    def __init__(self, obs_dim, act_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, act_dim)
        )

    def forward(self, x):
        return self.net(x)

policy = BCPolicy(obs_dim=observations.shape[1], act_dim=actions.shape[1])
optimizer = optim.Adam(policy.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()

# -----------------------------
# Training Loop with loss tracking
# -----------------------------
num_epochs = 20
loss_history = []

for epoch in range(num_epochs):
    epoch_loss = 0
    for obs_batch, act_batch in loader:
        pred = policy(obs_batch)
        loss = loss_fn(pred, act_batch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item() * obs_batch.size(0)
    epoch_loss /= len(tensor_dataset)
    loss_history.append(epoch_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.6f}")

# -----------------------------
# Test the learned policy in the environment
# -----------------------------
obs, _ = env.reset()
done = False
total_reward = 0

while not done:
    obs_tensor = torch.tensor(obs, dtype=torch.float32).unsqueeze(0)
    action = policy(obs_tensor).detach().numpy()[0]
    obs, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated
    total_reward += reward

print(f"Total reward from BC policy: {total_reward}")

```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import matplotlib.pyplot as plt
import minari

# Define multiple BC architectures
class BCPolicySmall(nn.Module):
    def __init__(self, obs_dim, act_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, act_dim)
        )
    def forward(self, x):
        return self.net(x)

class BCPolicyMedium(nn.Module):
    def __init__(self, obs_dim, act_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, act_dim)
        )
    def forward(self, x):
        return self.net(x)

class BCPolicyLarge(nn.Module):
    def __init__(self, obs_dim, act_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, act_dim)
        )
    def forward(self, x):
        return self.net(x)

# Model configs
architectures = {
    "Small (128x128)": BCPolicySmall,
    "Medium (256x256)": BCPolicyMedium,
    "Large (512x512x256)": BCPolicyLarge
}

# Parameters
num_epochs = 20
batch_size = 64
learning_rate = 1e-3
dataset_names = [
    "mujoco/halfcheetah/simple-v0",
    "mujoco/halfcheetah/medium-v0",
    "mujoco/halfcheetah/expert-v0"
]

# Store results
dataset_results = {}

for dataset_name in dataset_names:
    print(f"\n=== Dataset: {dataset_name} ===")
    dataset = minari.load_dataset(dataset_name, download=True)
    env = dataset.recover_environment()

    # Prepare data
    episodes = dataset.sample_episodes(dataset.total_episodes)
    obs_list, act_list = [], []
    for ep in episodes:
        obs_list.append(ep.observations[:-1])
        act_list.append(ep.actions)
    observations = torch.tensor(np.concatenate(obs_list, axis=0), dtype=torch.float32)
    actions = torch.tensor(np.concatenate(act_list, axis=0), dtype=torch.float32)
    tensor_dataset = TensorDataset(observations, actions)
    loader = DataLoader(tensor_dataset, batch_size=batch_size, shuffle=True)

    loss_histories = {}
    rewards = {}

    for arch_name, arch_class in architectures.items():
        print(f"\n--- Training {arch_name} ---")
        policy = arch_class(obs_dim=observations.shape[1], act_dim=actions.shape[1])
        optimizer = optim.Adam(policy.parameters(), lr=learning_rate)
        loss_fn = nn.MSELoss()

        loss_history = []
        for epoch in range(num_epochs):
            epoch_loss = 0
            for obs_batch, act_batch in loader:
                pred = policy(obs_batch)
                loss = loss_fn(pred, act_batch)
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item() * obs_batch.size(0)
            epoch_loss /= len(tensor_dataset)
            loss_history.append(epoch_loss)
            print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.6f}")

        # Save loss history
        loss_histories[arch_name] = loss_history

        # Test policy
        obs, _ = env.reset()
        done = False
        total_reward = 0
        while not done:
            obs_tensor = torch.tensor(obs, dtype=torch.float32).unsqueeze(0)
            action = policy(obs_tensor).detach().numpy()[0]
            obs, reward, terminated, truncated, info = env.step(action)
            done = terminated or truncated
            total_reward += reward

        rewards[arch_name] = total_reward
        print(f"{arch_name} Reward: {total_reward}")

    # Store results for dataset
    dataset_results[dataset_name] = {
        "losses": loss_histories,
        "rewards": rewards
    }

    # Plot loss curves for this dataset
    plt.figure(figsize=(8, 5))
    for arch_name, losses in loss_histories.items():
        plt.plot(range(1, num_epochs + 1), losses, marker='o', label=arch_name)
    plt.xlabel("Epoch")
    plt.ylabel("MSE Loss")
    plt.title(f"Training Loss - {dataset_name}")
    plt.legend()
    plt.grid(True)
    plt.show()

# Summary: Print rewards for all datasets & architectures
print("\n=== Summary of Rewards ===")
for dataset_name, results in dataset_results.items():
    print(f"\nDataset: {dataset_name}")
    for arch_name, reward in results["rewards"].items():
        print(f"  {arch_name}: {reward:.2f}")
```

```python
# W2-3 — Custom TD3+BC (λ from Q(s,a) w/ ensemble-mean; targets updated on actor steps) — EXACT

import numpy as np, torch, torch.nn as nn, torch.nn.functional as F
import random

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42
np.random.seed(SEED); random.seed(SEED); torch.manual_seed(SEED)

# === Pull transitions via d3rlpy's transition_picker (v2.x) ===
tp = dataset.transition_picker
episodes = dataset.episodes

# Infer dims
first_tr = tp(episodes[0], 0)
obs_dim = first_tr.observation.shape[0]
act_dim = first_tr.action.shape[0]

# --- StandardObservationScaler stats (eps=1e-3) ---
EPS = 1e-3
total_count = sum(ep.transition_count for ep in episodes)

obs_sum = np.zeros(obs_dim, dtype=np.float64)
for ep in episodes:
    for i in range(ep.transition_count):
        tr = tp(ep, i)
        obs_sum += tr.observation.astype(np.float64)
obs_mean = (obs_sum / max(1, total_count)).astype(np.float32)

sq_sum = np.zeros(obs_dim, dtype=np.float64)
for ep in episodes:
    for i in range(ep.transition_count):
        tr = tp(ep, i)
        diff = tr.observation.astype(np.float64) - obs_mean.astype(np.float64)
        sq_sum += diff * diff
obs_std = np.sqrt(sq_sum / max(1, total_count)).astype(np.float32)
obs_std = np.maximum(obs_std, EPS)

def norm_np(x: np.ndarray) -> np.ndarray:
    return (x - obs_mean) / obs_std

# --- Make flat replay arrays ---
S  = np.empty((total_count, obs_dim), dtype=np.float32)
A  = np.empty((total_count, act_dim), dtype=np.float32)
R  = np.empty((total_count, 1), dtype=np.float32)
SN = np.empty((total_count, obs_dim), dtype=np.float32)
TER= np.empty((total_count, 1), dtype=np.float32)

k = 0
for ep in episodes:
    for i in range(ep.transition_count):
        tr = tp(ep, i)
        S[k]      = norm_np(tr.observation.astype(np.float32))
        A[k]      = tr.action.astype(np.float32)
        R[k, 0]   = float(tr.reward)     # safe scalar cast
        SN[k]     = norm_np(tr.next_observation.astype(np.float32))
        TER[k, 0] = float(tr.terminal)   # true env terminal; timeouts will bootstrap
        k += 1
assert k == total_count
num_transitions = total_count

# --- Action bounds (per-dimension clamp) ---
act_high = np.asarray(eval_env.action_space.high, dtype=np.float32)
act_low  = np.asarray(eval_env.action_space.low,  dtype=np.float32)
act_limit = float(np.max(np.abs(act_high)))
low_t  = torch.from_numpy(act_low).to(device)
high_t = torch.from_numpy(act_high).to(device)

# --- Torch tensors on device ---
S_t  = torch.from_numpy(S).to(device)
A_t  = torch.from_numpy(A).to(device)
R_t  = torch.from_numpy(R).to(device)
SN_t = torch.from_numpy(SN).to(device)
TER_t= torch.from_numpy(TER).to(device)

print(f"Custom dataset tensors: transitions={num_transitions}, obs_dim={obs_dim}, act_dim={act_dim}, act_limit={act_limit}")

# =========================
# Networks (TD3+BC)
# =========================
def mlp(in_dim, out_dim):
    return nn.Sequential(
        nn.Linear(in_dim, 256), nn.ReLU(),
        nn.Linear(256, 256), nn.ReLU(),
        nn.Linear(256, out_dim)
    )

class Actor(nn.Module):
    def __init__(self, obs_dim, act_dim, act_limit):
        super().__init__()
        self.net = mlp(obs_dim, act_dim)
        self.act_limit = act_limit
    def forward(self, s):
        return torch.tanh(self.net(s)) * self.act_limit

class Critic(nn.Module):  # twin Q
    def __init__(self, obs_dim, act_dim):
        super().__init__()
        self.q1 = mlp(obs_dim + act_dim, 1)
        self.q2 = mlp(obs_dim + act_dim, 1)
    def forward(self, s, a):
        sa = torch.cat([s, a], dim=-1)
        return self.q1(sa), self.q2(sa)
    def q1_only(self, s, a):
        return self.q1(torch.cat([s, a], dim=-1))

actor = Actor(obs_dim, act_dim, act_limit).to(device)
actor_targ = Actor(obs_dim, act_dim, act_limit).to(device)
critic = Critic(obs_dim, act_dim).to(device)
critic_targ = Critic(obs_dim, act_dim).to(device)
actor_targ.load_state_dict(actor.state_dict())
critic_targ.load_state_dict(critic.state_dict())

# =========================
# Hyperparameters (match d3rlpy defaults)
# =========================
lr = 3e-4
pi_opt = torch.optim.Adam(actor.parameters(), lr=lr)
q_opt  = torch.optim.Adam(critic.parameters(), lr=lr)
gamma = 0.99
tau = 0.005
policy_delay = 2
target_noise_std = 0.2
target_noise_clip = 0.5
alpha = 2.5
mse = nn.MSELoss()

@torch.no_grad()
def soft_update(src, dst, tau):
    for p, p_t in zip(src.parameters(), dst.parameters()):
        p_t.data.mul_(1 - tau).add_(tau * p.data)

# ---- Target-update cadence (TD3 style) ----
TARGET_UPDATE_MODE = 'delayed'  # update targets only when actor updates

# =========================
# Training
# =========================
total_steps = 5_000
batch_size = 256
log_every = 1_000
rng = np.random.default_rng(123)

actor.train(); critic.train()
running = {"critic": 0.0, "actor": 0.0, "bc": 0.0, "cnt_actor_updates": 0, "cnt_steps": 0}

for step in range(1, total_steps + 1):
    idx = rng.integers(0, num_transitions, size=batch_size)
    s  = S_t[idx]; a = A_t[idx]; r = R_t[idx]; sn = SN_t[idx]; d = TER_t[idx]

    # ----- Critic update -----
    with torch.no_grad():
        noise = (torch.randn_like(a) * target_noise_std).clamp(-target_noise_clip, target_noise_clip)
        a2 = actor_targ(sn) + noise
        a2 = torch.max(torch.min(a2, high_t), low_t)   # per-dimension clamp
        q1_t, q2_t = critic_targ(sn, a2)
        q_targ = torch.min(q1_t, q2_t)                 # TD3 target reduction = 'min'
        y = r + gamma * (1.0 - d) * q_targ             # bootstrap at timeouts

    q1, q2 = critic(s, a)
    critic_loss = mse(q1, y) + mse(q2, y)
    q_opt.zero_grad(set_to_none=True)
    critic_loss.backward()
    q_opt.step()

    if TARGET_UPDATE_MODE == 'every':
        soft_update(critic, critic_targ, tau)
        soft_update(actor,  actor_targ,  tau)

    # ----- Delayed actor update -----
    # ----- Delayed actor update -----
    if step % policy_delay == 0:
        # π(s) and Q(s,π(s)) must require grad
        pi_s = actor(s)                            # [B, act_dim]
        q1_pi, q2_pi = critic(s, pi_s)             # twin critics on π(s)
        q_pi_mean = 0.5 * (q1_pi + q2_pi)          # actor uses ensemble mean (q_func_factory='mean')

        # ===== IMPORTANT: λ from dataset actions Q(s, a) (detach), ensemble-mean =====
        with torch.no_grad():                      # no grad through λ
            q1_da, q2_da = critic(s, a)
            q_da_mean = 0.5 * (q1_da + q2_da)
            lam = alpha / q_da_mean.abs().mean().clamp_min(1e-12)   # scalar

        # Actor loss = -λ * E[Q_mean(s, π(s))] + BC MSE
        bc_loss = ((a - pi_s) ** 2).mean()
        actor_loss = (-lam * q_pi_mean.mean()) + bc_loss

        pi_opt.zero_grad(set_to_none=True)
        actor_loss.backward()
        pi_opt.step()

        if TARGET_UPDATE_MODE == 'delayed':
            # Match d3rlpy: update targets ***after*** the actor step only
            soft_update(critic, critic_targ, tau)
            soft_update(actor,  actor_targ,  tau)

        running["actor"] += float(actor_loss.item())
        running["bc"]    += float(bc_loss.item())
        running["cnt_actor_updates"] += 1

    running["critic"] += float(critic_loss.item())
    running["cnt_steps"] += 1

    if step % log_every == 0 or step == total_steps:
        c_steps = max(1, running["cnt_steps"])
        a_steps = max(1, running["cnt_actor_updates"])
        print(f"[{step}/{total_steps}] critic_loss={running['critic']/c_steps:.3f}  "
              f"actor_loss={running['actor']/a_steps:.3f}  bc_loss≈{running['bc']/a_steps:.3f}")
        running = {"critic": 0.0, "actor": 0.0, "bc": 0.0, "cnt_actor_updates": 0, "cnt_steps": 0}

# Save checkpoint with scaler + bounds
torch.save({
    "actor": actor.state_dict(),
    "critic": critic.state_dict(),
    "obs_mean": torch.from_numpy(obs_mean).float(),
    "obs_std": torch.from_numpy(obs_std).float(),
    "act_limit": act_limit,
    "act_low": torch.tensor(act_low),
    "act_high": torch.tensor(act_high),
}, "td3bc_custom_light.pt")
print("Saved: td3bc_custom_light.pt")
```

```python
# [W2-4] — Manual evaluation (strictly matching d3rlpy scaling, dtype & shape)

import numpy as np
import torch
import torch.nn as nn
import d3rlpy

# ---- Ensure eval env is available ----
try:
    eval_env
except NameError:
    _, eval_env = d3rlpy.datasets.get_minari("mujoco/halfcheetah/simple-v0")

# ---- Reference (d3rlpy) algorithm: use in-memory 'algo' if present, else load ----
try:
    td3bc_ref = algo
except NameError:
    td3bc_ref = d3rlpy.load_learnable("td3bc_d3rlpy_light.d3", device="cpu")

# ---- Pull the EXACT fitted StandardObservationScaler from d3rlpy ----
obs_scl = getattr(td3bc_ref, "observation_scaler", None)
assert obs_scl is not None and getattr(obs_scl, "built", False), "d3rlpy observation_scaler not found or not built."
ref_mean = np.asarray(obs_scl.mean, dtype=np.float32)
ref_std  = np.maximum(np.asarray(obs_scl.std, dtype=np.float32), getattr(obs_scl, "eps", 1e-3))

# ---- Our custom actor & (our) saved scaler (for comparison) ----
ckpt = torch.load("td3bc_custom_light.pt", map_location="cpu")
our_mean = ckpt["obs_mean"].numpy() if isinstance(ckpt["obs_mean"], torch.Tensor) else ckpt["obs_mean"].astype(np.float32)
our_std  = ckpt["obs_std"].numpy()  if isinstance(ckpt["obs_std"],  torch.Tensor) else ckpt["obs_std"].astype(np.float32)
act_limit = float(ckpt["act_limit"])

obs_dim = eval_env.observation_space.shape[0]
act_dim = eval_env.action_space.shape[0]
act_low  = eval_env.action_space.low.astype(np.float32)
act_high = eval_env.action_space.high.astype(np.float32)

# ---- Rebuild our Actor (exactly as trained) ----
class Actor(nn.Module):
    def __init__(self, obs_dim, act_dim, act_limit):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 256), nn.ReLU(),
            nn.Linear(256, 256), nn.ReLU(),
            nn.Linear(256, act_dim)
        )
        self.act_limit = act_limit
    def forward(self, s):
        return torch.tanh(self.net(s)) * self.act_limit

actor = Actor(obs_dim, act_dim, act_limit)
actor.load_state_dict(ckpt["actor"])
actor.eval()

# ---- Greedy action functions ----
def act_ref(obs_np: np.ndarray) -> np.ndarray:
    a = td3bc_ref.predict(obs_np.astype(np.float32).reshape(1, -1))[0]
    return np.clip(a, act_low, act_high)

@torch.no_grad()
def act_custom_with_scaler(obs_np: np.ndarray, mean: np.ndarray, std: np.ndarray) -> np.ndarray:
    x = (obs_np.astype(np.float32) - mean) / std
    a = actor(torch.from_numpy(x).unsqueeze(0)).squeeze(0).numpy()
    return np.clip(a, act_low, act_high)

# ---- Shared rollout (identical seeds for both) ----
def run_policy_env(env, act_fn, n_episodes=5, seed=123):
    rets = []
    for ep in range(n_episodes):
        obs, _ = env.reset(seed=seed + ep)
        done = False; ret = 0.0
        while not done:
            a = act_fn(obs)
            obs, r, term, trunc, _ = env.step(a)
            ret += float(r)
            done = term or trunc
        rets.append(ret)
    return np.array(rets, dtype=np.float32)

N_EVAL = 5
ref_returns   = run_policy_env(eval_env, act_ref, n_episodes=N_EVAL, seed=123)
cust_refscale = run_policy_env(eval_env, lambda o: act_custom_with_scaler(o, ref_mean, ref_std), n_episodes=N_EVAL, seed=123)
cust_ourscale = run_policy_env(eval_env, lambda o: act_custom_with_scaler(o, our_mean, our_std), n_episodes=N_EVAL, seed=123)

def summarize(arr):
    m = float(arr.mean())
    s = float(arr.std(ddof=1) if arr.size > 1 else 0.0)
    return m, s

m_ref,  s_ref  = summarize(ref_returns)
m_refS, s_refS = summarize(cust_refscale)
m_ourS, s_ourS = summarize(cust_ourscale)

print("\n==== HalfCheetah (Minari) — TD3+BC Comparison (5 eps) ====")
print(f"Reference (d3rlpy):     mean={m_ref:.2f},  std={s_ref:.2f},  returns={np.round(ref_returns, 2)}")
print(f"Custom (ref scaler):    mean={m_refS:.2f}, std={s_refS:.2f}, returns={np.round(cust_refscale, 2)}")
print(f"Custom (our  scaler):   mean={m_ourS:.2f}, std={s_ourS:.2f}, returns={np.round(cust_ourscale, 2)}")

# Scaler sanity: how different are the vectors?
print("\nScaler Δ (∞-norm):  mean:", float(np.max(np.abs(ref_mean - our_mean))), " std:", float(np.max(np.abs(ref_std - our_std))))
```

```python
# W2-6c — Manual RGB capture to MP4 for Reference (d3rlpy) and Custom TD3+BC

# If you already ran W2-6a to set OSMesa, you can skip re-installing.
# Otherwise, uncomment these:
# import os
# os.environ["MUJOCO_GL"] = "osmesa"
# os.environ["PYOPENGL_PLATFORM"] = "osmesa"
!pip -q install imageio imageio-ffmpeg

import os, numpy as np, torch, torch.nn as nn
import imageio.v2 as imageio
import d3rlpy
import gymnasium as gym
from IPython.display import HTML, display

# --- Make an RGB env (same spec as dataset)
def make_rgb_env():
    try:
        return dataset.recover_environment(render_mode="rgb_array")
    except Exception:
        return gym.make("HalfCheetah-v4", render_mode="rgb_array")

# --- Load reference algo (use in-memory 'algo' if present)
try:
    td3bc_ref = algo
except NameError:
    td3bc_ref = d3rlpy.load_learnable("td3bc_d3rlpy_light.d3", device="cpu")

# --- Load our custom actor + scaler
ckpt = torch.load("td3bc_custom_light.pt", map_location="cpu")
obs_mean = ckpt["obs_mean"].numpy() if isinstance(ckpt["obs_mean"], torch.Tensor) else ckpt["obs_mean"]
obs_std  = ckpt["obs_std"].numpy()  if isinstance(ckpt["obs_std"],  torch.Tensor) else ckpt["obs_std"]
act_limit = float(ckpt["act_limit"])

# Build a temp env to get spaces for actor and clipping
_tmp = make_rgb_env()
obs_dim = _tmp.observation_space.shape[0]
act_dim = _tmp.action_space.shape[0]
act_low  = _tmp.action_space.low.astype(np.float32)
act_high = _tmp.action_space.high.astype(np.float32)
horizon  = getattr(_tmp.spec, "max_episode_steps", 1000)
_tmp.close()

# Rebuild the custom Actor exactly as trained
class Actor(nn.Module):
    def __init__(self, obs_dim, act_dim, act_limit):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 256), nn.ReLU(),
            nn.Linear(256, 256), nn.ReLU(),
            nn.Linear(256, act_dim)
        )
        self.act_limit = act_limit
    def forward(self, s):
        return torch.tanh(self.net(s)) * self.act_limit

actor = Actor(obs_dim, act_dim, act_limit)
actor.load_state_dict(ckpt["actor"])
actor.eval()

# Action functions
def act_ref(obs_np: np.ndarray) -> np.ndarray:
    a = td3bc_ref.predict(obs_np.astype(np.float32).reshape(1, -1))[0]
    return np.clip(a, act_low, act_high)

@torch.no_grad()
def act_custom(obs_np: np.ndarray) -> np.ndarray:
    x = (obs_np.astype(np.float32) - obs_mean) / obs_std
    a = actor(torch.from_numpy(x).unsqueeze(0)).squeeze(0).numpy()
    return np.clip(a, act_low, act_high)

# Manual MP4 writer
def run_episode_to_mp4(act_fn, mp4_path, seed=123, fps=30, max_steps=None):
    env = make_rgb_env()
    obs, _ = env.reset(seed=seed)
    total_reward, steps = 0.0, 0
    horizon_local = max_steps or horizon

    frames_written = 0
    writer = imageio.get_writer(mp4_path, fps=fps, codec="libx264")

    # write initial frame
    frame = env.render()
    if frame is not None:
        writer.append_data(frame); frames_written += 1

    for t in range(horizon_local):
        a = act_fn(obs)
        obs, r, term, trunc, _ = env.step(a)
        total_reward += float(r); steps += 1

        frame = env.render()
        if frame is not None:
            writer.append_data(frame); frames_written += 1

        if term or trunc:
            break

    writer.close()
    env.close()

    if frames_written == 0:
        raise RuntimeError("No frames were captured; check MuJoCo GL backend (OSMesa) and render_mode='rgb_array'.")
    return total_reward, steps, mp4_path, frames_written

# Record both videos
ref_ret, ref_steps, ref_path, ref_frames = run_episode_to_mp4(act_ref,    "halfcheetah_ref.mp4", seed=123)
cus_ret, cus_steps, cus_path, cus_frames = run_episode_to_mp4(act_custom, "halfcheetah_custom.mp4", seed=123)

print(f"Reference: return={ref_ret:.2f}, steps={ref_steps}, frames={ref_frames}, video={ref_path}")
print(f"Custom   : return={cus_ret:.2f}, steps={cus_steps}, frames={cus_frames}, video={cus_path}")

# Display side-by-side
display(HTML(f"""
<div style="display:flex;gap:16px;align-items:flex-start;">
  <div>
    <h4 style="margin:4px 0;">Reference (d3rlpy) — return {ref_ret:.1f}</h4>
    <video width="420" controls src="{ref_path}"></video>
  </div>
  <div>
    <h4 style="margin:4px 0;">Custom — return {cus_ret:.1f}</h4>
    <video width="420" controls src="{cus_path}"></video>
  </div>
</div>
"""))
```
